```python
# üìâ Add Loss Ratio column
df['LossRatio'] = df['TotalClaims'] / df['TotalPremium']
df[['TotalClaims', 'TotalPremium', 'LossRatio']].head()
```

```python
# üìä Histogram of Total Premium
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
sns.histplot(df['TotalPremium'], kde=True)
plt.title("Distribution of Total Premium")
plt.xlabel("TotalPremium")
plt.ylabel("Frequency")
plt.grid(True)
plt.tight_layout()
plt.savefig("total_premium_hist.png")
plt.show()
```

```python
# üìå Average Loss Ratio by Province
plt.figure(figsize=(10,5))
df.groupby('Province')['LossRatio'].mean().sort_values().plot(kind='bar', color='teal')
plt.title("Average Loss Ratio by Province")
plt.ylabel("Average Loss Ratio")
plt.xlabel("Province")
plt.tight_layout()
plt.savefig("loss_ratio_by_province.png")
plt.show()
```

```python
# üì¶ Box Plot: Outliers in Total Claims
plt.figure(figsize=(8,5))
sns.boxplot(x=df['TotalClaims'], color='salmon')
plt.title("Boxplot of Total Claims")
plt.tight_layout()
plt.savefig("boxplot_total_claims.png")
plt.show()
```
